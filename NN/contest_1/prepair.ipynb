```python
import os
import shutil
from tqdm import trange
src_dir = '/AkhmetzyanovD/projects/MTS/NN/contest_1/nn-image/train_test_data/train'
dest_dir = '/AkhmetzyanovD/projects/MTS/NN/contest_1/images_check'
for i in trange(len(data)):
    df_i = data.iloc[i]
    dest_subdir = os.path.join(dest_dir, df_i['label'])
    if not os.path.exists(dest_subdir):
        os.makedirs(dest_subdir)

    src_path = os.path.join(src_dir, df_i['filename'])
    dest_path = os.path.join(dest_subdir, df_i['filename'])
    shutil.copy(src_path, dest_path)
```

```python
mapper = {name: i for i, name in enumerate(data['label'].unique())}
mapper
```

```python
data['label'] = data['label'].map(mapper)
data
```

```python
data.to_csv('/AkhmetzyanovD/projects/MTS/NN/contest_1/nn-image/train_val.csv', index=False)
```

```python
import os
import numpy as np
from PIL import Image
```

```python
img_dir = '/AkhmetzyanovD/projects/mts_contest_2/nn-image/train_test_data'
paths = []
```

```python
for part in ('train', 'test'):
    for img in os.listdir(os.path.join(img_dir, part)):
        paths.append(os.path.join(img_dir, part, img))
```

```python
rgb_values = np.concatenate(
    [Image.open(img).getdata() for img in paths], 
    axis=0
) / 255.

mu_rgb = np.mean(rgb_values, axis=0)
std_rgb = np.std(rgb_values, axis=0)
```

```python
import cv2

img_path = '/AkhmetzyanovD/projects/MTS/NN/contest_1/nn-image/train_test_data/train/0a00fe3cd103394e092e4190d2e9d010cbbe649de55939e8f5c5f053b86d52f2.jpg'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
import albumentations as A

train_transforms = A.Compose([
        A.Resize(height=224, width=224, interpolation=cv2.INTER_LINEAR, p=1.0),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0),
        A.ShiftScaleRotate(
            shift_limit=(0, 0), 
            scale_limit=(-0, 0), 
            rotate_limit=(0, 0),
            border_mode=1
            ),
        A.PixelDropout(0.08, p=1),
        A.CoarseDropout(10, 16, 16, 0, 1, 1)
        
        ])
```

```python
trm_img = train_transforms(image=img)['image']
```
